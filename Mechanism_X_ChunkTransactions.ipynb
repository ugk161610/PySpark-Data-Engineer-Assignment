```python
from pyspark.sql.functions import monotonically_increasing_id

df = spark.read.option("header", True).csv("abfss://9e108cca-f00a-4ed8-a107-b025c1d99e01@onelake.dfs.fabric.microsoft.com/5e7beabe-a64d-478b-a5ec-47cccc4ea716/Files/raw_data/transactions.csv")
```

```python
df = df.withColumn("row_id", monotonically_increasing_id())
```

```python
chunk_size = 10000
total_rows = df.count()
num_chunks = (total_rows // chunk_size) + 1
```

```python

for i in range(num_chunks):
    chunk_df = df.filter((df.row_id >= i * chunk_size) & (df.row_id < (i + 1) * chunk_size))
    chunk_df.drop("row_id").write.mode("overwrite").parquet(f"Files/chunks/chunk_{i}.parquet")
```
